```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
# Import a sample binary outcome train/test set into H2O
train = h2o.import_file("https://s3.amazonaws.com/erin-data/higgs/higgs_train_10k.csv")
test = h2o.import_file("https://s3.amazonaws.com/erin-data/higgs/higgs_test_5k.csv")
```

```python
train.head()
```

```python
# Identify predictors and response
x = train.columns
y = "response"
x.remove(y)
```

```python

# For binary classification, response should be a factor
train[y] = train[y].asfactor()
test[y] = test[y].asfactor()
```

```python
import time 
```

```python
start=time.perf_counter() 
# Run AutoML for 20 base models
aml = H2OAutoML(max_models=20, seed=1)
aml.train(x=x, y=y, training_frame=train)
end=time.perf_counter() 
```

```python
print('total time {}'.format(end-start))
```

```python
# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf

import autokeras as ak
```

```python
train_file_path = tf.keras.utils.get_file("train.csv", 'https://s3.amazonaws.com/erin-data/higgs/higgs_train_10k.csv')
test_file_path = tf.keras.utils.get_file("eval.csv", 'https://s3.amazonaws.com/erin-data/higgs/higgs_test_5k.csv')
```

```python
import time 
start=time.perf_counter() 
# Initialize the structured data classifier.
clf = ak.StructuredDataClassifier(
    overwrite=True, max_trials=3
)  # It tries 3 different models.
# Feed the structured data classifier with training data.
clf.fit(
    # The path to the train.csv file.
    train_file_path,
    # The name of the label column.
    "response",
    epochs=20,
)
end=time.perf_counter()
print('total time: {}'.format(end-start))
```

```python
# Predict with the best model.
predicted_y = clf.predict(test_file_path)
# Evaluate the best model with testing data.
print(clf.evaluate(test_file_path, "response"))
```
